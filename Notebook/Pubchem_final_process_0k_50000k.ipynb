```python
sol_new=data1.Solubility1
import re
sol1=[]
### Using regular expression module cleaning the data and taking out the data with with unit "mg/mL" in the pubchem ...
for s in sol_new:
   #sol=re.findall(r'(?<![\-\/])(\d+(?:\.?\d+)) *mg\b', s)
   sol=re.findall('(?<![\-\/])(\d+(?:\.?\d+)).*( M|ug/mL|mg/mL|(ug/mL)|mg/L|ml)', s)
   sol1.append(sol)
print(len(sol1))
```

```python
import pandas as pd



# Creating a new DataFrame df3 with SMILES, CID, and Solubility1
df3 = data1[['SMILES', 'CID', 'Solubility1']].reset_index(drop=True)

# Creating df2 DataFrame
df2 = pd.DataFrame(sol1, columns=['Solubility2'])

# Concatenating df3 and df2
df_new = pd.concat([df3, df2], axis=1, ignore_index=False)

# Renaming the columns of df_new
df_new.columns = ['SMILES', 'CID', 'Solubility1', 'Solubility2']

# Dropping rows where 'Solubility2' is NaN
df_sol = df_new.loc[df_new['Solubility2'].notnull()]

# Resetting the index of df_sol
df_sol1 = df_sol.reset_index(drop=True)

# Printing the final DataFrame
df_sol1
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# Function to calculate molecular weight from a SMILES string
def calculate_molecular_weight_from_smiles(smiles):
    try:
        molecule = Chem.MolFromSmiles(smiles)
        if molecule is None:
            raise ValueError(f"Invalid SMILES string: {smiles}")
        molecular_weight = Descriptors.MolWt(molecule)
        return molecular_weight
    except Exception as e:
        print(f"Error processing SMILES '{smiles}': {e}")
        return None  



# Apply the function to the 'SMILES' column to calculate molecular weight
df_sol1['Molecular_Weight'] = df_sol1['SMILES'].apply(calculate_molecular_weight_from_smiles)


```

```python
df_sol1
```

```python
df2=pd.DataFrame()
for sol in df_sol1.Solubility2:
  #df1=pd.DataFrame()
  df=pd.DataFrame(sol)
  df=df.T
  df1=df.iloc[:,0:2]
  #print(df)
  #df5=pd.concat([df1,df],axis=0)
  df2=pd.concat([df2,df1],axis=0,ignore_index=True)
df_final=pd.concat([df_sol1,df2],axis=1)
df_final.rename(columns={ df_final.columns[5]: "sol_new1" }, inplace = True)
df_final.rename(columns={ df_final.columns[6]: "Unit" }, inplace = True)
df_final.sol_new1 = pd.to_numeric(df_final.sol_new1, errors="coerce")
df_final['sol_final']=''
for i in range (len(df_final)):

  if df_final['Unit'][i]=='ug/mL':
    df_final['sol_final'][i]=(df_final['sol_new1'][i]/1000).round(4)
  elif df_final['Unit'][i]=='mg/mL':
    df_final['sol_final'][i]=df_final['sol_new1'][i]
  elif df_final['Unit'][i]=='mg/L':
    df_final['sol_final'][i]=(df_final['sol_new1'][i]/1000).round(4)
  elif df_final['Unit'][i]=='ml':
    df_final['sol_final'][i]=df_final['sol_new1'][i]
  elif df_final['Unit'][i]=='mg/ml':
    df_final['sol_final'][i]=(df_final['sol_new1'][i]).round(4)    
  elif df_final['Unit'][i]==' M':
    df_final['sol_final'][i]=(df_final['sol_new1'][i] * df_final['Molecular_Weight'][i]).round(4)
df_final
```

```python
#### Function to convert smiles to canonical smiles 

from rdkit.Chem import MolFromSmiles as smi2mol
from rdkit.Chem import MolToSmiles as mol2smi
## Function to create canonical smiles 
def canon(smi):
    try:
        mol=smi2mol(smi, sanitize=True)
        smi_canon=mol2smi(mol, isomericSmiles=False, canonical=True)
        #smi_canon=mol2smi(mol, isomericSmiles=True,canonical=True) ### According to Sorkun 
        return(smi_canon)
    except:
        print("ERROR")
        return(smi)
```

```python
### Applying function to create the column with canonical smiles.  
df1['smiles_canon'] = [canon(smi) for smi in df1.SMILES]
df2['smiles_canon'] = [canon(smi) for smi in df2.smiles_canon]
```

```python
### Finding the matching rows train and test data based on canonical smiles(Unique representation pf the smiles )
Match_rows = pd.merge(df1, df2, on=['smiles_canon'], how='inner')
#Match_rows = pd.merge(df1, df2, on='smiles_canon', suffixes=('_pubchem', '_lit'), how='inner')

print(len(Match_rows))
#Match_rows
#Match_rows.to_csv('data/overlap_data.csv', index=False)
```
